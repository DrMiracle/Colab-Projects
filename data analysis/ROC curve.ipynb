# Ініціалізація функцій

```python
import numpy as np
import matplotlib.pyplot as plt

def my_roc_function(predicted_probabilities, true_labels):
    # Обчислюємо значення TPR та FPR для різних порогів передбаченої ймовірности
    thresholds = sorted(set(predicted_probabilities), reverse=True)
    tpr_values = []
    fpr_values = []

    for threshold in thresholds:
        prediction = predicted_probabilities >= threshold
        tp = np.sum((predicted_probabilities >= threshold) & true_labels)
        fp = np.sum((predicted_probabilities >= threshold) & ~true_labels)
        tn = np.sum((predicted_probabilities < threshold) & ~true_labels)
        fn = np.sum((predicted_probabilities < threshold) & true_labels)
        tpr = tp / (tp + fn)
        fpr = fp / (fp + tn)
        tpr_values.append(tpr)
        fpr_values.append(fpr)

    # Побудова кривої ROC
    plt.plot(fpr_values, tpr_values, 'b', label='ROC curve')
    plt.plot([0, 1], [0, 1], 'r--', label='Random guess')
    plt.title('ROC curve')
    plt.xlabel('False Positive Rate (FPR)')
    plt.ylabel('True Positive Rate (TPR)')
    plt.grid(True)
    plt.legend()
    plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import roc_curve, auc

def roc_function_with_lib(predicted_probabilities, true_labels):
    fpr_values, tpr_values, thresholds = roc_curve(true_labels, predicted_probabilities)
    roc_auc = auc(fpr_values, tpr_values)

    # Побудова кривої ROC
    plt.plot(fpr_values, tpr_values, 'b', label='ROC curve')
    plt.title('ROC curve with libraries')
    plt.xlabel('False Positive Rate (FPR)')
    plt.ylabel('True Positive Rate (TPR)')
    plt.grid(True)
    plt.legend()
    plt.show()
```
