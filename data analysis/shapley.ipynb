```python
import numpy as np
import matplotlib.pyplot as plt

# Define the profit calculation function
def calculate_profits(cost):
    firm_A_profit = (min(cost, 120) + max(0, cost - 230))/ 2
    firm_B_profit = (min(cost, 230) + max(0, cost - 120))/ 2
    return firm_A_profit, firm_B_profit

# Generate the range of costs
costs = np.linspace(40, 300, 261)  # 261 points from 40 to 300

# Calculate profits for each cost
firm_A_profits = []
firm_B_profits = []

for cost in costs:
    firm_A_profit, firm_B_profit = calculate_profits(cost)
    firm_A_profits.append(firm_A_profit)
    firm_B_profits.append(firm_B_profit)

# Convert lists to numpy arrays for easier plotting
firm_A_profits = np.array(firm_A_profits)
firm_B_profits = np.array(firm_B_profits)

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(costs, firm_A_profits, label='Firm A Shapley values')
plt.plot(costs, firm_B_profits, label='Firm B Shapley values')
plt.xlabel('Cost')
plt.ylabel('Profit')
plt.title('Shapley values of Firm A and Firm B vs. Cost')
plt.legend()
plt.grid(True)
plt.show()
```
