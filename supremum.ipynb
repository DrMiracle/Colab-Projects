```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0, 0.04, 1.95, 0.38, 0.20, 0.24, 1.14, 0.10, 1.96, 1.00, 0.07])
x = np.sort(x)
n = x.size
y = np.linspace(0, 1, n)

xx = np.array([x, x]).T.flatten()[1:]
yy = np.array([y, y]).T.flatten()[:-1]

expx = np.linspace(0,2,50)
expy = 1 - np.exp(-expx/1)

plt.plot(xx, yy, label='Fn(x)')
plt.plot(expx, expy, label='F(x,1)')
plt.plot(x, y, '.', color='black')
plt.legend()
plt.show()

sup = np.abs(y - (1 - np.exp(-x/1))).max()
print("Supremum: {}".format(sup))
```
